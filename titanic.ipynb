```python
import pandas as pd
import numpy as np
url="https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv"
df=pd.read_csv(url)
df
```

```python
gender=pd.get_dummies(df.Sex)
embark=pd.get_dummies(df.Embarked)
embark
```

```python
ij=pd.concat([df,gender],axis='columns')
kl=pd.concat([df,gender,embark],axis='columns')
kl.drop(["Name","Cabin","Sex","Ticket","Embarked","Survived"],axis="columns",inplace=True)
kl.Age=kl.Age.fillna(kl.Age.mean())
kl
```

```python
ij.drop("Sex",axis='columns',inplace=True)
```

```python
ij.drop("Ticket",axis='columns',inplace=True)
```

```python
ij.drop(["Name","Parch","SibSp","Cabin","Embarked"],axis="columns",inplace=True)
```

```python
ij
```

```python
ij.Age=ij.Age.fillna(ij.Age.mean())
```

```python
ij
```

```python
target=ij["Survived"]
```

```python
ij.drop("Survived",axis="columns",inplace=True)
```

```python
ij
```

```python
target
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ij,target,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
x_train1,x_test1,y_train1,y_test1=train_test_split(kl,target,test_size=0.2)
```

```python
model.fit(x_train1,y_train1)
```

```python
model.score(x_test1,y_test1)
```
