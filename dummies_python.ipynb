```python
import pandas as pd
import numpy as np
from sklearn import linear_model
url="https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv"
df=pd.read_csv(url)
df
```

```python
df.rename(columns={'Car Model':'carmodel','Sell Price($)':'sellprice','Age(yrs)':'age'},inplace=True)
df
```

```python
ij=pd.get_dummies(df.carmodel)
```

```python
ij
```

```python
kl=pd.concat([df,ij],axis='columns')
```

```python
kl
```

```python
kl.drop(['carmodel'],axis='columns')
```

```python
kl.rename(columns={'Mercedez Benz C class':'benz'},inplace=True)
```

```python
kl
```

```python
kl.drop('benz',axis='columns',inplace=True)
```

```python
kl
```

```python
kl.drop('carmodel',axis='columns',inplace=True)
```

```python
kl
```

```python
reg=linear_model.LinearRegression()
```
